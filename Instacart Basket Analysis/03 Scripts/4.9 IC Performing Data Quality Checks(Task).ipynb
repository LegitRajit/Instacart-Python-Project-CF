## Addressing Missing Values

We have a significant amount of missing First Name values. Upon further thought, both First Name and Last Name are both unneccessary as we can use user_id to identify each customer. I will drop both of these columns in the dataframe

## Checking for Mixed Type Data

```python
# Function for checking for mixed-type columns
for col in df_cstmer_2.columns.tolist():
  weird = (df_cstmer_2[[col]].applymap(type) != df_cstmer_2[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cstmer_2[weird]) > 0:
    print (col, ' mixed')
  else: print (col, ' consistent')
```

## Addressing Mixed Type Data

No Mixed Data was found after the check, so no further action needs to be taken.
