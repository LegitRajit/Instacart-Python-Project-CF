## Price Range Flag

```python
# Subsetting dataframe
df_op_merged_2 = df_ords_prods_m[:1000000]
```

```python
# Create Function "Price_Label" that separates items by price range
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Creating new column "price_range" and applying price_label function to it.
df_op_merged_2['price_range'] = df_op_merged_2.apply(price_label, axis = 1)
```

2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Updated For Loop to include the two busiest days and two slowest days of the week 
Activity = []

for value in df_ords_prods_m2["order_dow"]:
  if value == 0 or value == 1:
    Activity.append("Busiest Days")
  elif value == 3  or value == 4:
    Activity.append("Slowest Days")
  else:
    Activity.append("Regular Days")
```

# 05 Busiest Period of Day Flag

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day

## Creating Flag

```python
df_ords_prods_m2['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Loop to group Order Frequency based on hour 
Grouped_Hours = []
Most_Orders = [9,10,11,12,13,14,15,16]
Average_Orders = [7,8,17,18,19,20,21,22]
Fewest_Orders = [0,1,2,3,4,5,6,23]


for value in df_ords_prods_m2["order_hour_of_day"]:
  if value in Most_Orders:
    Grouped_Hours.append("Most_Orders")
  elif value in Fewest_Orders:
    Grouped_Hours.append("Fewest_Orders")
  else:
    Grouped_Hours.append("Average_Orders")
```

```python
Grouped_Hours
```

```python
#Creating new column for updated Hour Activity
df_ords_prods_m2['busiest_period_of_day'] = Grouped_Hours
```

```python
df_ords_prods_m2['busiest_period_of_day'].value_counts(dropna = False)
```

```python
df_ords_prods_m2.head()
```
